```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/plouzane-1758197_1280.jpg",0)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

#image in frequency domain

```python
img2=np.fft.fft2(img)
plt.imshow(np.log1p(np.abs(img2)),cmap='gray')
plt.axis('off')
plt.show()
```

# Shifiting low frequency component to centere of the image

```python
Fshift=np.fft.fftshift(img2)
plt.imshow(np.log1p(np.abs(Fshift)),cmap='gray')
plt.axis('off')
plt.show()
```

# Filter: Low pass filter

```python
M,N = img.shape
H = np.zeros((M,N), dtype=np.float32)
D0 = 50

for u in range(M):
  for v in range(N):
    D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
    if D <= D0:
      H[u,v] = 1
    else:
      H[u,v] = 0

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

# Ideal Low Pass Filtering

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)),
cmap='gray')
plt.axis('off')
plt.show()
```

# Inverse the shifted Fourier Transform

```python
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G))),
cmap=('gray')
plt.axis('off')
plt.show()
```

#Inverse the fourier transformation

```python
g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap="gray")
plt.axis("off")
plt.show()
```

# Filter: High pass filter

```python
H = 1 - H

plt.imshow(H, cmap="gray")
plt.axis("off")
plt.show()
```

# Ideal High Pass Filtering

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)),cmap="gray")
plt.axis("off")
plt.show()
```

# Inverse Fourier Transform

```python
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G)),cmap="gray")
plt.axis("off")
plt.show()

g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap="gray")
plt.axis("off")
plt.show()
```
